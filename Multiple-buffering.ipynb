```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('E:/PAPERS/#02/district.shp')
```

```python
gdf.plot()
```

```python
selected = gdf.sample(1)
```

```python
fig, ax = plt.subplots()
selected.buffer(2000,join_style=1).plot(ax=ax,color='red',alpha=0.5)
selected.buffer(4000,join_style=1).plot(ax=ax,color='red',alpha=0.5)
selected.buffer(6000,join_style=1).plot(ax=ax,color='red',alpha=0.5)


selected.plot(ax=ax,color='blue')
```
